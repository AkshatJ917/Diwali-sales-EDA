Importing the Libraries and  Data set

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as py
import seaborn as sb
df = pd.read_csv(r"C:\Users\Dell\Downloads\Diwali Sales Data.csv",encoding='latin1')
```

```python
df.head(20)
```

Data Cleaning

```python
#Dropping Useless Columns
```

```python
df1 = df.drop(columns=["Status","unnamed1"])
```

```python
df1.shape
```

```python
#Dropping rows which contains null values
```

```python
df1.dropna(inplace=True)
```

```python
df1.shape
```

```python
#Checking for Null values
```

```python
pd.isnull(df1).sum()
```

```python
df1.info()
```

```python
#Changing the data Type of Amount from float to int
```

```python
df1["Amount"]= df1["Amount"].astype(int)
```

```python
df1.info()
```

```python
#Removing all Duplicates
```

```python
df1 = df1.drop_duplicates()
```

```python
df1.duplicated().sum()
```

```python
#Changing the return value from Marital_Status column
```

```python
df1['Marital_Status'] = df1['Marital_Status'].map({0:'Married',1:'Non-Married'})
```

```python
df1
```

Analysis based on Gender

```python
ax = sb.countplot(x='Gender',hue = 'Gender',data = df1)
sb.set(rc={'figure.figsize':(7,5)})
ax.set_title('No. of Customers by Gender')
ax.set_ylabel('No. of customers')

for bars in ax.containers:
    ax.bar_label(bars)
```

```python
Sales_by_gen = df1.groupby(["Gender"], as_index=False)["Amount"].sum().sort_values(by="Amount",ascending=False)
```

```python
bx = sb.barplot(x='Gender',y='Amount',hue = 'Gender',data = Sales_by_gen )
sb.set(rc={'figure.figsize':(7,5)})
bx.set_title('Total Sales by Gender')
```

Analysis based on Age Group

```python
age_order =['0-17','18-25','26-35','36-45','46-50','51-55','55+']
cx = sb.countplot(x='Age Group',hue='Gender',data=df1,order = age_order)
sb.set(rc={'figure.figsize':(10,5)})
cx.set_title("Customer Age Group")
cx.set_ylabel("No. of customers")
```

```python
cust_sales = df1.groupby(["Age Group"], as_index=False)["Amount"].sum().sort_values(by="Amount",ascending=False)
```

```python
age_order =['0-17','18-25','26-35','36-45','46-50','51-55','55+']
dx = sb.barplot(x='Age Group',y='Amount',hue='Age Group',data=cust_sales,order=age_order)
sb.set(rc={'figure.figsize':(10,5)})
dx.set_title("Sales by Age Group")
dx.set_ylabel("Sales")
```

Analysis based on State

```python
order_state = df1.groupby(["State"], as_index=False)["Orders"].sum().sort_values(by ='Orders',ascending=False).head(10)
```

```python
ex = sb.barplot(x='State',y='Orders',hue='State',data= order_state)
sb.set(rc={'figure.figsize':(18,5)})
ex.set_title("Top 10 States by Orders")
ex.set_xlabel("States")
ex.set_ylabel("No. of orders")

for bars in ex.containers:
    ex.bar_label(bars)
```

```python
state_sales = df1.groupby(['State'], as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False).head(10)
```

```python
fx = sb.barplot(x='State',y='Amount',hue='State',data=state_sales)
sb.set(rc={'figure.figsize':(18,5)})
fx.set_title("Top 10 States by Sales")
fx.set_xlabel("States")
fx.set_ylabel("Sales")
```

Analysis based on Marital Status

```python
gx= sb.countplot(x='Marital_Status',hue='Marital_Status',data = df1)
sb.set(rc={'figure.figsize':(7,5)})
gx.set_title("Customers by Marital Status")
gx.set_xlabel("Marital Status")
gx.set_ylabel("No. of customers")

for bars in gx.containers:
    gx.bar_label(bars)
```

```python
marr_sales = df1.groupby(['Marital_Status','Gender'], as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
```

```python
hx = sb.barplot(x='Marital_Status',y='Amount',hue='Gender',data = marr_sales)
hx.set_title("Sales by Marital Status")
hx.set_xlabel("Marital Status")
hx.set_ylabel("Sales")
```

Analysis based on Occupation

```python
ix = sb.countplot(x='Occupation',hue='Occupation',data = df1,order=df['Occupation'].value_counts(ascending=False).index)
sb.set(rc={'figure.figsize':(21,7)})
ix.set_title("Customers By occupation")
ix.set_ylabel("No. of Customers")

for bars in ix.containers:
    ix.bar_label(bars)
```

```python
occu_sales = df1.groupby(['Occupation'], as_index=False)['Amount'].sum().sort_values(by = 'Amount',ascending=False)
```

```python
jx = sb.barplot(x='Occupation',y='Amount',hue='Occupation',data=occu_sales)
sb.set(rc={'figure.figsize':(21,7)})
jx.set_title("Sales by Occupation")
jx.set_ylabel("Sales")
```

Analysis based on Product Category

```python
prod_order = df1.groupby(['Product_Category'], as_index=False)['Orders'].sum().sort_values(by = 'Orders',ascending = False).head(10)
```

```python
lx = sb.barplot(x='Product_Category',y='Orders',hue='Product_Category',data=prod_order)
sb.set(rc={'figure.figsize':(20,5)})
lx.set_title("Top 10 product category based on orders")
lx.set_xlabel("Products")
lx.set_ylabel("No. of Orders")

for bars in lx.containers:
    lx.bar_label(bars)
```

```python
prod_sales = df1.groupby(['Product_Category'], as_index=False)['Amount'].sum().sort_values(by = 'Amount',ascending = False).head(10)
```

```python
mx = sb.barplot(x='Product_Category',y='Amount',hue='Product_Category',data=prod_sales)
sb.set(rc={'figure.figsize':(20,5)})
mx.set_title("Top 10 product category based on Sales")
mx.set_xlabel("Products")
mx.set_ylabel("Sales")
```

Conclusion:-
Married customers form the largest customer segment and contribute the highest overall sales, with married females being the strongest contributors.
The 26–35 age group dominates both customer count and spending, making it the most valuable demographic segment.
Uttar Pradesh, Maharashtra, and Karnataka emerge as the top-performing states in terms of customers and revenue contribution.
Customers working in the IT, Healthcare, and Aviation sectors generate the highest sales, indicating stronger purchasing power.
While clothing, food, and electronics are the most frequently ordered categories, food products alone drive the highest revenue contribution.
